```python
import cv2
import matplotlib.pyplot as plt
```

```python
#lectura de imagen en formato BGR (Blue, Green y Red)
img = cv2.imread('lena.jpg')
```

```python
plt.figure()
```

```python
plt.imshow(img)
```

```python
plt.show() # sólo necesario para los IDE, no para notebook.
```

```python
#transformación a niveles de gris, por herramienta cvtColor (BGR a RGB)
img_color_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(dpi=150)
```

```python
plt.imshow(img_color_rgb)
```

```python
#separación de canales
blue, green, red = cv2.split(img) # la variable img contiene 3 matrices, RGB. cuyo orden debe ser (azul, verde y roja)
```

```python
plt.figure()
plt.imshow(blue, cmap='gray') # la variable cmap es una paleta de colores
plt.show()
```

```python
plt.figure()
plt.imshow(red, cmap='gray') # la variable cmap es una paleta de colores
plt.show()
```

```python
plt.figure()
plt.imshow(green, cmap='gray') # la variable cmap es una paleta de colores
plt.show()
```

```python
print(blue.shape) #imprimir la dimensión de un canal.
```

```python
#calculo de histograma
hist_red = cv2.calcHist([img],[2],None,[256],[0,256]) # utilizando matriz red
```

```python
plt.figure()
plt.imshow(red, cmap='gray') # la variable cmap es una paleta de colores
plt.show()

plt.plot(hist_red)
plt.figure(dpi=100)
plt.show()
```

```python
#calculo de histograma
hist_green = cv2.calcHist([img],[1],None,[256],[0,256]) #utilizando matriz green
```

```python
plt.figure()
plt.imshow(green, cmap='gray') # la variable cmap es una paleta de colores
plt.show()

plt.plot(hist_green, label='green')
plt.figure(dpi=100)
plt.show()
```

```python
#calculo de histograma
hist_blue = cv2.calcHist([img],[0],None,[256],[0,256]) #utilizando matriz blue
```

```python
plt.figure()
plt.imshow(blue, cmap='gray') # la variable cmap es una paleta de colores
plt.show()

plt.plot(hist_blue)
plt.figure(dpi=100)
plt.show()
```

```python
plt.plot(hist_green, color='green', label='green')
plt.plot(hist_red, color='red',label='red')
plt.plot(hist_blue, color='blue',label='blue')
plt.legend()
plt.show()
```

```python
#*******************
# programa principal
#*******************

#lectura imagen
img = cv2.imread('cameraman.png')
gris = img[:,:,0]
norm = gris / 255

output = norm**5
output2 = norm**(1/4)

print(norm)
plt.imshow(norm, cmap='gray')
plt.show()

print(output)
plt.imshow(output, cmap='gray')
plt.show()

print(output2)
plt.imshow(output2, cmap='gray')
plt.show()
```

```python
#lectura de la imagen
lena= cv2.imread('lena.jpg')
barbara= cv2.imread('barbara_gray.bmp')

#transformación a niveles de gris
gray1 = cv2.cvtColor(lena, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(barbara, cv2.COLOR_BGR2GRAY)

#suma
suma = cv2.add(gray1,gray2)

plt.imshow(gray1, cmap="gray")
plt.show()

plt.imshow(gray2, cmap="gray")
plt.show()

plt.imshow(suma, cmap="gray")
plt.show()
```
