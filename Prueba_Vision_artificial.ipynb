### 4. Ecualice cada uno de los histogramas obtenidos en el paso 3 según el algoritmo visto en clases y muestre el resultado en una ventana para cada canal.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#lectura de la imagen
img1 = cv2.imread('img/fotocy.jpg')

#transformación a nivel de gris
src = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

#ecualizacion
dst = cv2.equalizeHist(src)

blue_1, green_1, red_1 = cv2.split(img1)
```

Inicialmente se produce la carga de la misma imagen para ser monocanalizada a tono de grises (referencia posterior), por otra parte, se efectúa un split de los 3 canales (BGR) y se aplica la ecualización a cada canal, tal como se muestra a continuación:

#### - Imagen ecualizada e histograma de la Variable **B (Blue)**:

```python
eq_blue  = cv2.equalizeHist(blue_1)
plt.figure()
plt.imshow(eq_blue, cmap='gray') # la variable cmap es una paleta de colores
plt.title("Fotografía Ecualizada, canal B")
plt.show()
print(eq_blue.shape) #imprimir la dimensión del canal Azul (Blue).

#calculo de histograma
hist_eq_blue = cv2.calcHist([eq_blue],[0],None,[256],[0,256]) # seleccionamos el canal 0 siempre (sólo hay 1 canal).
plt.plot(hist_eq_blue)
plt.title("Histograma Ecualizado, canal B")
plt.figure(dpi=100)
plt.show()
```

#### - Imagen ecualizada e histograma de la Variable **G (Green)**:

```python
eq_green = cv2.equalizeHist(green_1)
plt.figure()
plt.imshow(eq_green, cmap='gray') # la variable cmap es una paleta de colores
plt.title("Fotografía Ecualizada, canal G")
plt.show()
print(eq_green.shape) #imprimir la dimensión del canal green (verde)

#calculo de histograma
hist_eq_green = cv2.calcHist([eq_green],[0],None,[256],[0,256])
plt.plot(hist_eq_green)
plt.title("Histograma Ecualizado, canal G")
plt.figure(dpi=100)
plt.show()
```

#### - Imagen ecualizada e histograma de la Variable **R (Red)**:

```python
eq_red   = cv2.equalizeHist(red_1)
plt.figure()
plt.imshow(eq_red, cmap='gray') # la variable cmap es una paleta de colores
plt.title("Fotografía Ecualizada, canal R")
plt.show()
print(eq_red.shape) #imprimir la dimensión del canal rojo (red).

#calculo de histograma
hist_eq_red = cv2.calcHist([eq_red],[0],None,[256],[0,256])
plt.plot(hist_eq_red)
plt.title("Histograma Ecualizado, canal R")
plt.figure(dpi=100)
plt.show()
```

#### - Comparación de ambos histogramas (Original y Ecualizados):

```python
rows = 1
columns = 2
fig = plt.figure(figsize=(20, 5))
fig.add_subplot(rows, columns, 1)

plt.plot(hist_green, color='green', label='green')
plt.plot(hist_red, color='red',label='red')
plt.plot(hist_blue, color='blue',label='blue')
plt.legend()
plt.title("Histograma Original, 3 canales")

fig.add_subplot(rows, columns, 2)
plt.plot(hist_eq_green, color='green', label='eq.green')
plt.plot(hist_eq_red, color='red',label='eq.red')
plt.plot(hist_eq_blue, color='blue',label='eq.blue')
plt.legend()
plt.title("Histograma Ecualizado, 3 canales")
```

Como se observa en la figura, el histograma de frecuencia acumulada de grises en los 3 canales ecualizados demuestra una distribución más uniforme en el espectro (rango) que el histograma original, distribuyendo los niveles de grises uniformemente

#### - Comparación de ambas fotografías a color (Original y Ecualizada):

```python
m,n,ch = img1.shape

#seteamos el output para tres canales de tipo uint8
out =np.zeros((m,n,3), dtype='uint8')
out[:,:,0]= eq_red
out[:,:,1]= eq_green
out[:,:,2]= eq_blue

rows = 1
columns = 2
fig = plt.figure(figsize=(10, 7))
fig.add_subplot(rows, columns, 1)

plt.imshow(img1[:,:,[2,1,0]])
plt.title("Fotografía Original")

fig.add_subplot(rows, columns, 2)
plt.imshow(out)
plt.title("Fotografía Ecualizada")
```

### 5. Mejore la imagen ecualizada canal G obtenida en el paso 4, empleando la función gama con valores 1.5 y 0.4. Guarde estos resultados en distintos archivos

Objeto efectuar mejoramiento en el espacio, se pueden aplicar transformaciones básicas a niveles de grises como es el uso de la ecuación de curva gama. Cabe destacar que a mayor número de gama, la imagen se oscurece, por otra parte, cuando es menor a 1, las áreas oscuras se aclaran.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def gamma_correction(eq_green, factor):
	eq_green = eq_green/255.0
	eq_green = cv2.pow(eq_green, factor)
	return np.uint8(eq_green*255)
```

#### - Corrección de gama con factor 1.5 al canal G (Green) ecualizado.

```python
eq_green = cv2.equalizeHist(green_1)

rows = 1
columns = 2
fig = plt.figure(figsize=(10, 7))
fig.add_subplot(rows, columns, 1)
plt.imshow(eq_green, cmap='gray') # la variable cmap es una paleta de colores
plt.title("Foto Ecualizada, canal G")
print(eq_green.shape) #imprimir la dimensión del canal green (verde)

#llamado a funcion
output_15 = gamma_correction(eq_green, 1.5)

fig.add_subplot(rows, columns, 2)
plt.imshow(output_15)
plt.title("Foto Ecualizada - corrección gamma 1.5, canal G")
```

#### - Corrección de gama con factor 0.4 al canal G (Green) ecualizado.

```python
eq_green = cv2.equalizeHist(green_1)

rows = 1
columns = 2
fig = plt.figure(figsize=(10, 7))

fig.add_subplot(rows, columns, 1)
plt.imshow(eq_green, cmap='gray') # la variable cmap es una paleta de colores
plt.title("Foto Ecualizada, canal G")
print(eq_green.shape) #imprimir la dimensión del canal green (verde)

#llamado a funcion
output_04 = gamma_correction(eq_green, 0.4)

fig.add_subplot(rows, columns, 2)
plt.imshow(output04)
plt.title("Foto Ecualizada - corrección gamma 0.4, canal G")
```
